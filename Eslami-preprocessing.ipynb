```
from IPython import display
import cv2
import matplotlib.pyplot as plt
from skimage.io import imread
from skimage.color import rgb2gray
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from skimage import exposure
from skimage import data, img_as_float
from skimage import exposure
from skimage.feature import canny
from scipy import ndimage as ndi
from skimage import morphology
import scipy.misc
import scipy.ndimage
import skimage.filters
import sklearn.metrics
from PIL import Image, ImageFilter
from albumentations import MedianBlur
import imutils
from imutils import contours
PATH = '/content/drive/MyDrive/Eslami/test/IMD002.jpg'

original_img = cv2.imread(PATH, 0)

equ_img = cv2.equalizeHist(original_img)
plt.figure()
plt.figure(figsize=(10,15))
plt.imshow(equ_img, cmap='gray')
```

```
from scipy.ndimage import gaussian_filter
result = gaussian_filter(original_img, sigma=5)
plt.imshow(result,cmap='gray')
plt.show()
```

```
gray = cv2.Canny(np.uint8(original_img), 90, 100)
gray = cv2.dilate(gray, None, iterations=1)
gray = cv2.erode(gray, None, iterations=1)
plt.imshow(gray,cmap='gray')
plt.show()
```
