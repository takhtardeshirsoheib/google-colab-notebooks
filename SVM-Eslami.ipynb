```
from subprocess import check_output
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import numpy as np
import cv2
import os
image_dataset1 = "/content/drive/MyDrive/classification/melanoma/"
image_dataset2 = "/content/drive/MyDrive/classification/non-melanoma/"

list1 = os.listdir(image_dataset1)
list2 = os.listdir(image_dataset2)

for i in range (len(list1)):
  img = cv2.imread(image_dataset1 + list1[i])
  img = cv2.resize(img,(300,400))
  img = cv2.cvtColor(np.uint8(img), cv2.COLOR_BGR2RGB)
  plt.imsave('/content/drive/MyDrive/classification/resized-m/' + str(i)+'.jpg',img)
for j in range (len(list2)):
  img = cv2.imread(image_dataset2 + list2[j])
  img = cv2.resize(img,(300,400))
  img = cv2.cvtColor(np.uint8(img), cv2.COLOR_BGR2RGB)
  plt.imsave('/content/drive/MyDrive/classification/resized-n/' + str(j)+'.jpg',img)


```

```
arayeh = np.zeros(((len(list1)+len(list2)),300*400*3))
target = []

adr1 = '/content/drive/MyDrive/classification/resized-m/'
adr2 = '/content/drive/MyDrive/classification/resized-n/'
adr_list1 = os.listdir(adr1)
adr_list2 = os.listdir(adr2)
index =0

for i in range (len(list1)):
  img = cv2.imread(adr1 +adr_list1[i])
  target.append(1)
  img = img/127.5 - 1
  vec = np.ravel(img)
  arayeh[index]=vec
  index+=1

for i in range (len(list2)):
  img = cv2.imread(adr2 +adr_list2[i])
  target.append(0)
  img = img/127.5 - 1
  vec = np.ravel(img)
  arayeh[index]=vec
  index+=1

print(arayeh)
```

```
print(target)
```

```
from sklearn.model_selection import GridSearchCV, train_test_split
X_train, X_test, y_train, y_test = train_test_split(arayeh, np.array(target), test_size=0.3,random_state=109)
```

```
from sklearn import svm, metrics, datasets
from sklearn.utils import Bunch
clf = svm.SVC()
clf.fit(X_train, y_train)
```

```
y_pred = clf.predict(X_test)



print("Classification report for - \n{}:\n{}\n".format(
    clf, metrics.classification_report(y_test, y_pred)))
```
